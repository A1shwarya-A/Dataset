```python
# 'data' is dataset, check for null values
nulls = data.isnull()

# Count the number of null values in each column
null_counts = {}
for col in data.columns:
    if data[col].isnull().any():
        null_counts[col] = data[col].isnull().sum()
print(null_counts)
```

```python
import warnings
warnings.filterwarnings('ignore')

# Preprocess height
data['Height'] = data['Height'].str.replace("'", ".")
data['Height'] = pd.to_numeric(data['Height'], errors='coerce')
height_mode = data['Height'].mode()[0]
data['Height'].fillna(height_mode, inplace=True)

# Preprocess weight
data['Weight'] = data['Weight'].str.replace("lbs", "")
data['Weight'] = pd.to_numeric(data['Weight'], errors='coerce')
weight_mode = data['Weight'].mode()[0]
data['Weight'].fillna(weight_mode, inplace=True)

# Univariate analysis for height
plt.figure(figsize=(8, 6))
sns.boxplot(data=data['Height'], orient='h', palette='Set2')
plt.title('Boxplot of Height')
plt.xlabel('Height (in feet)')
plt.show()

# Univariate analysis for weight
plt.figure(figsize=(8, 6))
sns.boxplot(data=data['Weight'], orient='h', palette='Set3')
plt.title('Boxplot of Weight')
plt.xlabel('Weight (in lbs)')
plt.show()

```

```python
for col in data.columns:
    if data[col].dtype == 'float64' or data[col].dtype == 'int64':
        data[col].fillna(data[col].mean(), inplace=True)
data.head(2)
```

#### Weight:

### **Reason for selecting the chart type:**
Boxplots were chosen for both height and weight because they effectively display the distribution of continuous variables and identify potential outliers. Boxplots provide information about the central tendency, variability, and presence of outliers in the data.

### **Pre-attentive attributes used:**
1. Length: The length of the box and whiskers in the boxplot provides information about the spread and variability of the data.
2. Position: The position of the median line within the box helps identify the central tendency of the data distribution.

### **Gestalt principles used:**
1. Proximity: Outliers that fall beyond the whiskers of the boxplot are visually separated from the main distribution, facilitating their identification.

Another question that could help in the EDA of the given dataset is:

Question: How are the skill attributes (E.g., 'Crossing', 'Finishing', 'Dribbling', etc.) distributed among players with different 'Preferred Foot'?

To answer this question, we can create a grouped bar plot that shows the average skill attribute values for players with left and right preferred foot.

```python
# Filter out columns related to skill attributes and 'Preferred Foot'
skill_attributes = ['Crossing', 'Finishing', 'Dribbling', 'ShortPassing', 'Volleys',
                    'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
                    'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
                    'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
                    'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
                    'Marking', 'StandingTackle', 'SlidingTackle']
attributes = ['Preferred Foot'] + skill_attributes
filtered_data = data[attributes]

# Calculate average skill attribute values for players with left and right preferred foot
avg_attributes = filtered_data.groupby('Preferred Foot').mean().reset_index()

# Reshape the data for plotting
avg_attributes_melted = avg_attributes.melt(id_vars='Preferred Foot', var_name='Skill Attribute', value_name='Average Value')

# Create a grouped bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Skill Attribute', y='Average Value', hue='Preferred Foot', data=avg_attributes_melted, palette='Set2')
plt.title('Average Skill Attribute Values for Players with Left and Right Preferred Foot')
plt.xlabel('Skill Attribute')
plt.ylabel('Average Value')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Preferred Foot', loc='upper right')
plt.grid(axis='y')
plt.show()
```

##**Summary:**
The grouped bar plot shows the average skill attribute values for players with left and right preferred foot.
We can observe if there are any differences in skill levels between players with different preferred foot, providing insights into the importance of foot preference in player skills.

##**Reason for selecting the chart type:**
A grouped bar plot is chosen because it allows us to compare the average skill attribute values between players with left and right preferred foot for multiple attributes simultaneously.
The grouped bar plot effectively shows the differences or similarities in skill levels between the two groups, making it easy to compare the skill attributes across different foot preferences.

##**Pre-attentive attributes used:**
Length: The length of each bar represents the average skill attribute value, making it easy to compare the values between the two groups.
Color: Different colors are used for bars representing players with left and right preferred foot, aiding in distinguishing between the groups.

##**Gestalt principles used:**
Similarity: Bars of the same color (representing the same preferred foot) are perceived as belonging to the same group, facilitating comparison within each group.
Closure: The bars are grouped together visually, creating a sense of completeness and making it easier to compare the two groups as a whole.
